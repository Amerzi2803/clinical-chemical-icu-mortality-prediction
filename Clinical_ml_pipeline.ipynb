```python
X = df.drop("icu_mortality", axis=1)
y = df["icu_mortality"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
y_train.value_counts(normalize=True)
y_test.value_counts(normalize=True)
```

```python
print("TRAIN:\n", y_train.value_counts(normalize=True))
print("\nTEST:\n", y_test.value_counts(normalize=True))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_features = ['age','weight','glucose','creatinine','hemoglobin','wbc','platelets',
                'sodium','potassium','lactate']

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[num_features] = scaler.fit_transform(X_train[num_features])
X_test_scaled[num_features] = scaler.transform(X_test[num_features])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score


lr = LogisticRegression(max_iter=1000, random_state=42)

lr.fit(X_train_scaled, y_train)


y_pred = lr.predict(X_test_scaled)
y_proba = lr.predict_proba(X_test_scaled)[:,1]

```

```python

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nROC-AUC:", roc_auc_score(y_test, y_proba))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score


rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42,
    class_weight='balanced'  # important for medical imbalance
)


rf.fit(X_train, y_train)


y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)[:,1]
```

```python
cm_rf = confusion_matrix(y_test, y_pred_rf)
print("Confusion Matrix:\n", cm_rf)
```

```python
cr_rf = classification_report(y_test, y_pred_rf)
print("\nClassification Report:\n", cr_rf)
```

```python
roc_auc_rf = roc_auc_score(y_test, y_proba_rf)
print("\nROC-AUC:", roc_auc_rf)
```
